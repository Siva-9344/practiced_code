```python
'''
import numpy as np
rng = np.random.default_rng(42)          # reproducible randomness

'''

import numpy as np

rng = np.random.default_rng(42) # Normally, random numbers change every time you run the code. 
# But when you use a seed number (like 42), you get the same random values every time you run it. (But you can use any number: 1, 7, 2024, etc.)

print(rng.random(5)) # its create the float number
print()
print(rng.integers(0,10,size = 5)) # it tells cretea the random rumbers from 0 to 10 but generate 5 numbers only
print()

# np.percentile()
heights_cm = np.random.normal(168,10,1000) # it tells create the random 1000 fake heights in cm, avg height is 168, std dev 10 its mean +_ come around 10.
#print(heights_cm)

q25,q50,q75 = np.percentile(heights_cm,[25,50,75])
print('quartile 25: ',q25)
print('quartile 50: ',q50)
print('quartile 75: ',q75)
print('-'*20)
print()

'''
âš¡ Mini-tasks
Return the 10th, 90th and 99th percentile of heights_cm.

Given daily sales for a store, find the inter-quartile range (IQR = P75â€“P25).

Load the built-in sklearn.datasets.load_boston() (or any CSV of house prices) and report the median price per suburb.
'''

# Return the 10th, 90th and 99th percentile of heights_cm.

q10,q90,q99 = np.percentile(heights_cm,[10,90,99])
print('percentile q10  :',q10)
print('percentile q90  :',q90)
print('percentile q99  :',q99)
```

```python
# Given daily sales for a store, find the inter-quartile range (IQR = P75â€“P25).

# poision distribution is used because (poission is the counting process like sales in 1 hour, sales in one day, sales in 1 min also)
sales = rng.poisson(120,365) # its creates value like using poission 
# (rng.poisson(120,365))===> its mean 365 values create 120 sales perday poision mena +or-10 values is will come.

#print(sales)

q25,q50,q75 = np.percentile(sales,[25,50,75])
print('percentile 25 : ',q25)
print('percentile 50 : ',q50)
print('percentile 75 : ',q75)

IQR = q75-q25 
print(IQR)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Random number generator
rng = np.random.default_rng()

# Create 500 values with low spread (std = 0.5)
low_spread = rng.lognormal(mean=5, sigma=0.5, size=500)

# Create 500 values with high spread (std = 2)
high_spread = rng.lognormal(mean=5, sigma=2, size=500)

# Plotting both
plt.figure(figsize=(12, 5))

# Low spread plot
plt.subplot(1, 2, 1)
plt.hist(low_spread, bins=30, color='skyblue', edgecolor='black')
plt.title('Lognormal (mean=5, std=0.5)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# High spread plot
plt.subplot(1, 2, 2)
plt.hist(high_spread, bins=30, color='salmon', edgecolor='black')
plt.title('Lognormal (mean=5, std=2)')
plt.xlabel('Value')

plt.tight_layout()
plt.show()
```
