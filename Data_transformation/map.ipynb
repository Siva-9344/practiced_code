```python
import pandas as pd
file_path = r"C:\Users\Sivaraman\Desktop\study\data analytics\practice\files_from_df.xlsx"
df = pd.read_excel(file_path)
print(df)
```

```python
def categorized(score):
    if score > 800:
        return 'Excellent'
    elif score > 600:
        return 'Good'
    else:
        return 'Poor'
df['categorized_credit_score'] = df['CreditScore'].map(categorized)   
df[['Name','CreditScore','categorized_credit_score'  ]]
```

```python
# converting gender to numericgender male:1 , female:0
df['numeric_gender'] = df['Gender'].str.lower().map({'male':1, 'female':0})
print(df[['Name','Gender','numeric_gender']])
```

```python
def age_category(age):
    if age < 30:
        return 'Young'
    elif 30 <= age < 50:
        return 'Middle_Age'
    else:
        return 'Senior'
df['Age_group'] = df['Age'].map(age_category)
df[['Name', 'Age', 'Age_group']]



    
```

```python
# classifying salary level

df['Salary_Level'] = df['Salary'].map(lambda X: 'High' if X>100000 else 'Low')
print(df[['Name', 'Age', 'Salary', 'Salary_Level']])
```

```python
''''country_map = {'France': 1, 'Spain': 2, 'Germany': 3}
df['CountryID'] = df['Country'].map(country_map)
print(df[['Name', 'Country', 'CountryID']])'''

# country map
country_map = {'France':1, 'Germany':2, 'Spain':3}
df['Countrys_Id'] = df['Country'].map(country_map)
print(df[['Name', 'Country', 'Countrys_Id']])


```

```python
df['Monthly_Salary'] = df['Salary'].apply(lambda x: round(x/12, 2))
df[['Name', 'Age', 'Salary', 'Monthly_Salary']]
```

```python
'''df['FullName'] = df.apply(lambda row: row['Name'] + ' (' + row['Country'] + ')', axis=1)
print(df[['Name', 'Country', 'FullName']])
'''
df['FullName'] = df.apply(lambda x: x['Name'] +'/ '+ x['Country'], axis = 1)
print(df[['Name', 'Country', 'FullName']])
```

```python
print(df)
```

```python
''' doubt why not use df for acesse in function.(❌ This 'df' df['CreditScore'] will compare all values at once, not row by row!

Instead, using r['CreditScore'] ensures we check just one row at a time.

'''

def credit_category(r): # you put anything inside of bracket like 'r', 'row', anything...
    if r['CreditScore'] > 750 and r['Salary'] > 100000:
        return 'VIP'
    elif r['CreditScore'] > 650 and r['Salary'] > 50000 :
        return 'Premium'
    else:
        return 'Regular'
df['Credit_Category'] = df.apply(credit_category, axis = 1) # axis = 1 is mentioned row wise
print(df[['Name', 'CreditScore', 'Credit_Category' ]])
```

```python
# Calculating Loan Eligibility
def loan_eligibility(row):
    return row['Salary']*0.4 if row['CreditScore'] > 750 else row['Salary']*0.2  # 0.4 mean 40% per, 0.2 mean 20% per
df['Loan_Amount'] = df.apply(loan_eligibility, axis = 1)
print(df[['Name', 'Age', 'Salary', 'CreditScore', 'Loan_Amount']])
```

```python
df_summary = df[['CreditScore', 'Salary', 'Age']].apply(lambda x: [x.mean(), x.min(), x.max()])
df_summary.index = ['Mean', 'Min', 'Max']
print(df_summary)
```
