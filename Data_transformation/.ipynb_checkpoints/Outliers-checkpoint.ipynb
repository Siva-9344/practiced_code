```python
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = max(Q1 - 1.5 * IQR, 0)  # Avoid negative salaries
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)]
correct_mean_salary = df_cleaned['Salary'].mean()
print(correct_mean_salary)
```
