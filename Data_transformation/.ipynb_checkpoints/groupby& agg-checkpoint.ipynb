```python
import pandas as pd
file_path = r"C:\Users\Sivaraman\Desktop\study\data analytics\Data (for practice)\bank_details_withdate.xlsx"
df = pd.read_excel(file_path)
print(df)
```

```python
#Basic Grouping and Aggregation
# Basic Grouping and Aggregation

#Total Balance for Each Country
df['Total_bal'] = df.groupby('Country')['Balance'].transform('sum')
print(df)
```

```python
# Average Salary for Each Gender in Each Country
#avg_salary = df.groupby(['Country', 'Gender'])['Salary'].mean().reset_index()
#print(avg_salary)

df_grouped = df.groupby(['Country', 'Gender'], as_index = False)['Salary'].mean()
print(df_grouped)
```

```python
# Customers with More Than 2 Products & Their Average Balance
filtered_cus =df[ df['NumOfProducts'] > 2].groupby('CustomerId')['Balance'].mean().reset_index()
print(filtered_cus)
```

```python
#Minimum & Maximum Credit Score for Each Country

'''credit_score_stats = df.groupby('Country')['CreditScore'].agg(['min', 'max']).reset_index()
print(credit_score_stats)'''

cred_score_stas = df.groupby('Country')['CreditScore'].agg(['min','max']).reset_index()
print(cred_score_stas)
```

```python
#Percentage of Total Salary for Each Customer in Their Country
'''df['Total_Salary_Country'] = df.groupby('Country')['Salary'].transform('sum')
df['Salary_Percentage'] = (df['Salary'] / df['Total_Salary_Country']) * 100
print(df[['CustomerId', 'Country', 'Salary', 'Salary_Percentage']])
'''

df['Total_Sal_coun'] = df.groupby('Country')['Salary'].transform('sum')
df['Sal_per'] = (df['Salary']/ df['Total_Sal_coun'])
df['Sal_per'] = df['Sal_per']* 100
print(df[['CustomerId', 'Name', 'Salary', 'Sal_per']])
```

```python
#Cumulative Balance Per Customer
df['Cumsum_bal'] = df.groupby('CustomerId')['Balance'].cumsum()
print(df[['CustomerId', 'Salary', 'Cumsum_bal']])
```

```python
#sorted balnce
Sorted_bal = df.groupby('CustomerId')['Balance'].sum().reset_index().sort_values(by = 'Balance', ascending = False)
print(Sorted_bal)
```

```python
# Finding Outliers: Customers with High Salary (Above 100K)
'''high_salary_customers = df[df['Salary'] > 100000].groupby('CustomerId')['Salary'].agg(['count', 'mean']).reset_index()
print(high_salary_customers)
'''
high_Sal_cus = df[df['Salary'] > 100000].groupby('CustomerId')['Salary'].agg(['count', 'mean']).reset_index()
print(high_Sal_cus)
```

```python
import pandas as pd
file_path = r"C:\Users\Sivaraman\Desktop\study\data analytics\Data (for practice)\bank_details_withdate.xlsx"
df = pd.read_excel(file_path)
print(df)
```

```python
#  Most Important groupby() + agg() Transformations
#‚úÖ Basic Aggregation
#  Average salary per department
df.groupby('Dep')['Salary'].mean()
```

```python
# Min, max, mean salary per department
df.groupby('Dep')['Salary'].agg(['min','max','mean'])
```

```python
#Multiple Columns with Multiple Aggregations
#  Salary ‚Üí mean, sum and Experience ‚Üí min, max per department
df.groupby('Dep').agg({'Salary': ['mean', 'sum'],
                      'CreditScore': ['min', 'max']})

```

```python
df.groupby('Country').agg({'Age': ['min', 'max'], 
                          'Balance': ['min', 'max']})
```

```python
df.groupby('Gender').agg({'Salary': ['min', 'max'],
                         'Balance': ['min', 'max']})
```

```python

df.groupby('Dep').agg({'Salary': ['min', 'max']})
```

```python
# Salary range (max - min) per department
df.groupby('Dep')['Salary'].agg(lambda x : x.max() - x.min())
```

```python
#  üÜî Rename Columns After Aggregation
df.groupby('Dep')['Salary'].agg(Avg_Salary = 'mean', Salary_Sum = 'sum')
```

```python
#üîÑ Reset Index (for Clean Output)
df.groupby('Dep')['Salary'].agg('mean').reset_index()
```

```python
#Count number of employees in each department
df.groupby('Dep').size()
```

```python
#‚ùì Using .count() vs .size()
# Count non-null salary values per department
df.groupby('Dep')['Salary'].count()
```

```python
df.groupby('Dep').count()
```

```python
# Average salary by department and experience level
df.groupby(['Dep', 'Tenure'])['Salary'].agg(avg_sal = 'mean')
```

```python
# Apply Different Aggregations to Different Columns
# Mean of Salary and sum of Experience per Department

df.groupby('Dep').agg({'Salary': 'mean',
                      'Tenure': 'sum'})
```

```python
#Sort Grouped Results
# Top-paying departments by average salary

df.groupby('Dep')['Salary'].agg('mean').sort_values(ascending = False)
```

```python
#  Get Row with Max Salary in Each Group (its mean full row details)
# Get employee with highest salary in each department

df.loc[df.groupby('Dep')['Salary'].idxmax()]

#.idxmax() is a Pandas function used on a Series or DataFrame to find the index (or label) of the maximum value.
```

```python
# Groupby on Index
#  Useful when your grouping column is already an index

df.set_index('Dep').groupby('Dep')['Salary'].mean()
```

```python
df.set_index('Dep').groupby('Dep')['Salary'].mean()
```

```python
df.groupby('Dep').mean(numeric_only = True)
```
