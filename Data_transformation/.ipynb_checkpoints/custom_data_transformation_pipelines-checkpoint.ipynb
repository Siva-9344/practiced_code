```python
#  PRACTICE SET 1: Students Score Data
import pandas as pd

data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', None],
    'Math': [78, 85, None, 92, 88, 90],
    'Science': ['89', 'not available', '91', '85', '87', '95'],
    'Passed': ['yes', 'no', 'yes', 'yes', '', 'yes']
}

df = pd.DataFrame(data)

'''
ðŸ”§ Task:
Drop rows with missing student name.

Fill missing Math score with average.

Convert Science to numeric (handle errors).

Convert Passed column: 'yes' â†’ 1, 'no' â†’ 0, others â†’ NaN.

Drop rows with any NaN values.
'''
print(df)

df = df[df['Student'].notnull()]  # Drop rows with missing student name.

df['Math'] = pd.to_numeric(df['Math'], errors = 'coerce') # Fill missing Math score with average.
df['Math'] = df['Math'].fillna(df['Math'].mean())

df['Science'] = pd.to_numeric(df['Science'], errors = 'coerce') # Convert Science to numeric (handle errors).


df['Passed'] = df['Passed'].replace({'yes': 1, 'no':0})  # Convert Passed column: 'yes' â†’ 1, 'no' â†’ 0, others â†’ NaN.
df['Passed'] = pd.to_numeric(df['Passed'], errors = 'coerce')

df = df.dropna()

print()
print('After task :')
print()
print(df)
```

```python
# PRACTICE SET 2: Products and Prices

data = {
    'Product': ['Pen', 'Notebook', 'Eraser', 'Pencil', 'Pen', ''],
    'Price': ['10', '25', None, '15', 'ten', '5'],
    'Stock': [50, None, 20, 30, 60, 10]
}

df = pd.DataFrame(data)
print(df)
print()
'''
ðŸ”§ Task:
Remove rows with empty Product.

Convert Price to numeric.

Fill missing Price with median price.

Fill missing Stock with 0.

Remove duplicate Products (keep first).
'''

df = df[df['Product'].str.strip() != ''] # Remove rows with empty Product.
df = df[df['Product'].notnull()]

price_maping = {'ten' : 10} # Convert Price to numeric.
df['Price'] = df['Price'].replace(price_maping) 
df['Price'] = pd.to_numeric(df['Price'], errors = 'coerce')

df['Price'] = df['Price'].fillna(df['Price'].median()) # Fill missing Price with median price.

df['Stocks'] = df['Stock'].fillna(0)

print('After task: ')
print()
print(df)


```

```python
# âœ… PRACTICE SET 3: Customer Orders

data = {
    'Customer': ['John', 'Sara', 'Ali', 'Nina', 'Ali'],
    'Order_Amount': ['200', '150', 'NA', '300', '250'],
    'Order_Date': ['2024-02-10', '2024-03-15', 'wrong-date', '2024-01-25', '2024-03-18']
}

df = pd.DataFrame(data)
print(df)

'''
ðŸ”§ Task:
Convert Order_Amount to numeric, set invalids to NaN.

Convert Order_Date to datetime (set invalids to NaT).

Drop duplicate customers (keep latest order).

Add a new column: Month extracted from Order_Date.
'''
print()
print('After task: ')
df['Order_Amount'] = pd.to_numeric(df['Order_Amount'], errors = 'coerce') # task : Convert Order_Amount to numeric, set invalids to NaN.

df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors = 'coerce' )

df = df.sort_values('Order_Date' ,ascending = False).drop_duplicates('Customer', keep = 'first') # Drop duplicate customers (keep latest order)

'''df = df.sort_values('Order_Date', ascending=False).drop_duplicates('Customer', keep='first')
print(df)'''

df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors = 'coerce')
df['Month'] = df['Order_Date'].dt.month
print(df)
```

```python
# âœ… PRACTICE SET 4: Employees Data

data = {
    'Name': ['Aman', 'Rita', '', 'Neha', None],
    'Salary': [50000, 60000, 55000, None, 52000],
    'Department': ['HR', 'Finance', 'HR', 'IT', 'Finance'],
    'Joining_Date': ['2021-06-01', 'not available', '2020-05-15', '2022-01-01', '2019-08-10']
}




'''
ðŸ”§ Task:
Drop rows with blank/None names.

Fill missing salary with average salary.

Convert Joining_Date to datetime.

Filter only employees from â€˜Financeâ€™ department.
'''
df = pd.DataFrame(data)
print(df)
print()
print()
print('After task: ')
print()





df = df[df['Name'].notnull() & (df['Name'].str.strip() != '')] #Drop rows with blank/None names.

df['Salary'] = df['Salary'].fillna(df['Salary'].mean()) # Fill missing salary with average salary.

df['Date_time'] = pd.to_datetime(df['Joining_Date'], errors = 'coerce') #Convert Joining_Date to datetime.

filters = df[df['Department'] == 'Finance']  #Filter only employees from â€˜Financeâ€™ department.

print(df)
print()
print()
print()
print('filter finance only: ')
print()
print(filters)
```

```python
# âœ… PRACTICE SET 5: Online Course Students

import pandas as pd

data = {
    'Name': ['Ankit', 'Ravi', 'Sneha', None, 'Meena', ''],
    'Course': ['Python', 'Data Science', 'ML', 'AI', 'Python', 'Data Science'],
    'Rating': ['4.5', '5', 'NaN', '4.0', 'three', '5.0'],
    'Completed': ['yes', 'no', '', 'yes', 'yes', 'no']
}

df = pd.DataFrame(data)
'''
ðŸ”§ Tasks:
Remove rows with missing or blank Name.

Convert Rating to float (handle invalids).

Replace '' and 'NaN' in Completed with 'no'.

Create a new column Is_Completed: yes â†’ 1, no â†’ 0.


'''
print(df)
print()
print('After :')
df = df[df['Name'].notna() & (df['Name'].str.strip() != '')] # notna mean Keeps rows where Name is NOT Na #Task: Remove rows with missing or blank Name.
df['Rating'] = pd.to_numeric(df['Rating'], errors = 'coerce') # Convert Rating to float (handle invalids).
df['Completed'] = df['Completed'].replace(['', 'Nan'], 'no') # Replace '' and 'NaN' in Completed with 'no'.
df['IS_Completed'] = df['Completed'].map({'yes': 1, 'no': 0}) # Create a new column Is_Completed: yes â†’ 1, no â†’ 0.


print(df)

```

```python
# âœ… PRACTICE SET 6: Social Media Ads Performance

data = {
    'Ad_ID': [101, 102, 103, 104, 105],
    'Clicks': ['100', '85', 'N/A', '120', '95'],
    'Impressions': [1000, 950, 800, None, 1000],
    'Cost': ['30.5', '28.0', 'invalid', '35', '25']
}

df = pd.DataFrame(data)
'''
ðŸ”§ Tasks:
Convert Clicks to numeric.

Convert Cost to numeric (handle 'invalid').

Fill missing Impressions with median.

Add new column CTR = (Clicks / Impressions) * 100.
'''
print(df)
print()
print('After :')

df['Clicks'] = pd.to_numeric(df['Clicks'], errors = 'coerce') # Convert Clicks to numeric.

df['Cost'] = pd.to_numeric(df['Cost'], errors = 'coerce') # Convert Cost to numeric (handle 'invalid').

df['Impressions'] = df['Impressions'].fillna(df['Impressions'].median()) # Fill missing Impressions with median.

df['CTR'] = (df['Clicks']/ df['Impressions']) * 100
df = df.dropna()


print(df)
```

```python
# PRACTICE SET 7: E-Commerce Returns

data = {
    'Customer': ['Kiran', 'Mona', 'Sameer', 'Mona', 'Arun'],
    'Returned': ['Yes', 'No', 'yes', 'No', 'YES'],
    'Refund_Amount': ['100', '', 'N/A', '50', '30']
}

df = pd.DataFrame(data)

'''
ðŸ”§ Tasks:
Standardize Returned column to all lowercase.

Convert Refund_Amount to float (handle empty, N/A).

Remove duplicate entries (keep first).

Filter only customers who returned items.
'''
print(df)
print()
print()
print('After :')

df['Returned'] = df['Returned'].str.lower() # Standardize Returned column to all lowercase.

df['Refund_Amount'] = pd.to_numeric(df['Refund_Amount'], errors = 'coerce') # Convert Refund_Amount to float (handle empty, N/A).

df = df.drop_duplicates(subset = ['Customer'], keep = 'first') # Remove duplicate entries (keep first).

df = df[df['Returned'].str.lower() == 'yes'] # Filter only customers who returned items.

print(df)
```

```python
# âœ… PRACTICE SET 8: Fitness Tracker

data = {
    'User': ['Vikram', 'Anu', 'John', 'Sara', ''],
    'Steps': ['10000', '9500', 'eight thousand', '11000', '7000'],
    'Calories_Burned': [300, 280, 270, None, 250],
    'Workout': ['Yes', 'No', 'Yes', '', 'Yes']
}

df = pd.DataFrame(data)

'''
ðŸ”§ Tasks:
Remove rows with missing User.

Convert Steps to numeric (handle words).

Fill missing Calories_Burned with average.

Replace empty string in Workout with 'No'.
'''
print(df)
print()
print()
print('After :')

df = df[df['User'].notna() & (df['User'].str.strip() != '')] # Remove rows with missing User.

df['Steps'] = pd.to_numeric(df['Steps'], errors = 'coerce') # Convert Steps to numeric (handle words).

df['Calories_Burned'] = df['Calories_Burned'].fillna(df['Calories_Burned'].mean()) # Fill missing Calories_Burned with average.

df['Workout'] = df['Workout'].replace([''], 'no') # Replace empty string in Workout with 'No'.
#df['Completed'] = df['Completed'].replace(['', 'Nan'], 'no')
print(df)
```

```python
# âœ… PRACTICE SET 9: City Weather Report
data = {
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Delhi', None],
    'Temperature_C': ['38', '33', 'thirty', '37', '35'],
    'Humidity': ['30%', '45%', '50%', '', '40%'],
    'Rainfall': [0, 12, 5, 7, 10]
}

df = pd.DataFrame(data)

'''
ðŸ”§ Tasks:
Remove duplicate cities (keep first).

Convert Temperature_C to numeric.

Remove % from Humidity and convert to int.

Drop rows with missing City or Temperature.
'''
print(df)
print()
print()
print('After :')

df = df.drop_duplicates(subset = ['City'], keep = 'first') # Remove duplicate cities (keep first).
df['Temperature_C'] = pd.to_numeric(df['Temperature_C'], errors = 'coerce') # Convert Temperature_C to numeric.
df['Humidity'] = df['Humidity'].str.replace('%', '').apply(pd.to_numeric, errors  = 'coerce') # Remove % from Humidity and convert to int.
df.dropna(subset = ['City', 'Temperature_C'], inplace = True) # Drop rows with missing City or Temperature.
# df.dropna(subset=['City', 'Temperature_C'], inplace=True)

print(df)

```
