```python
'''
Create a Product class with properties product_name, price, and quantity.
Add methods for calculating the total cost and checking if the product is in stock. and add some extra task
'''
class Product:
    def __init__(self, product_name, price, quantity, discount):
        self.product_name = product_name
        self.price = price
        self.quantity = quantity
        self.discount = discount
        
    def total_cost(self):
        print(f'Product totalcost : {self.price * self.quantity}')
    def check_stock(self):
        if self.quantity > 0:
            print(f'Stock is available : {self.quantity} items left')
        else:
            print('stock is not available')
    def apply_discount(self):    
        if self.discount > 0:
            self.price = self.price - ((self.price/100)*self.discount) # updated price after discount

product1 = Product('mobile', 25000, 20,0)
product2 = Product('lap', 50000, 0,0)
product3 = Product('modem', 3000, 2,0)
product4 = Product('spk',2000, 10, 10)

product1.total_cost()
product2.check_stock()
product3.check_stock()
product4.apply_discount()
print('After applied discount price : ',product4.price)

```

```python
# Task 1: Create and Manage Cart Items
'''
Create a Product class with the properties name, price, quantity.

Create a Cart class that allows:

Adding a product to the cart.

Removing a product from the cart by its name.

Displaying all products in the cart.

Calculating the total price of products in the cart.

Applying a 15% discount to the total price if it exceeds 30,000.
'''
```

```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
        
    def display(self):
        print(f'product: {self.name}, Price: {self.price}, Quantity: {self.quantity}')

class Cart:
    def __init__(self):
        self.cart_items = []
        
    def add_products(self, products):
        self.cart_items.append(products)
        print('Product is succesfully added')
        
    def remove_products(self, product_name):
        for product in self.cart_items:
            if product.name == product_name:
                self.cart_items.remove(product)
                print(f'{product_name} removed from the cart')
                break
        else:
            print(f'{product_name} is not found')
            
    def show_cartitems(self):       
        print('cart_contains :')
        for product in self.cart_items:
            product.display()
            
    def total_calculate(self):
        total = sum([product.price * product.quantity for product in self.cart_items ])
        if total > 30000:
            total = total * 0.85
            print('15% discount is aplied')
        return total    
product1 = Product('Book', 400, 20)
product2 = Product('Pen', 5, 100)
product3 = Product('lap', 45000, 15)
            
product1.display()
product2.display()
product3.display()

print()
print()

cart = Cart()
cart.add_products(product1)
cart.add_products(product2)
cart.add_products(product3)
print()
cart.total_calculate()
print()
cart.remove_products('Book')
cart.show_cartitems()         # its work perfectly
```

```python
#  Inventory and Order System â€“ Full Code

class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock
        
    def check_stock(self, quantity):
        return self.stock >= quantity
        
    def reduce_stock(self, quantity):
        if self.check_stock(quantity):
            self.stock = self.stock - quantity
            return True
        else:
            return False

class CartItem:
    def __init__(self, product, quantity):
        self.product = product
        self.quantity = quantity
    def get_item_total(self):
        return self.product.price * self.quantity


class Cart: # customer cart
    def __init__(self):
        self.items = [] # Store the products the customer wants to buy â€” not the shop's stock.

    def add_items(self, product, quantity):
        if product.check_stock(quantity):
            product.reduce_stock(quantity)  # thi line for reduce stock in shop after customer bought 
            cart_item = CartItem(product, quantity)
            self.items.append(cart_item)
            print(f"Added {quantity} x {product.name} to cart.")
        else:
            print(f"not enough stock for {product.name}. Available: {product.stock}, Requested : {quantity}")

    def display_cart(self):
        print('\nCart Summary: ')
        for item in self.items:
            print(f"{item.product.name} - Quantity : {item.quantity},"
                 f"Total : {item.get_item_total():.2f}")
        print(f"\nFinal Cart Total: {self.get_total_amount():.2f}\n")

    def get_total_amount(self):
        return sum(item.get_item_total()for item in self.items)


    def save_cart_to_file(self, filename = 'cart_summary.txt'):
        with open(filename, 'w') as f:
            f.write('Cart Summary: \n')
            for item in self.items:
                f.write(f"{item.product.name} - Quantity : {item.quantity},"
                f"Unit_price : {item.product.price:.2f},"
                f"Total: {item.get_item_total():.2f}\n")
        print(f"Cart saved to {filename}.")
        
#Next Task: Add a Simple User System (Multiple Users & Their Carts)
class User:
    def __init__ (self, name):
        self.name = name
        self.cart = Cart()
        
    def  add_to_cart(self, product, quantity):
        print(f"\n{self.name} is adding to cart: ")
        self.cart.add_items(product, quantity)
        
    def view_cart(self):    
        print(f"\n{self.name}'s cart")
        self.cart.display_cart()
    def save_cart(self):
        filename = f"{self.name}_cart_summary.txt"
        self.cart.save_cart_to_file(filename)
        

if __name__ == "__main__":
    mobile = Product('Mobile', 20000, 10)
    laptop = Product('laptop', 50000, 5)
    speaker = Product('Speaker', 1500, 20)
    
    user1 = User('Alice')
    user2 = User('Bob')
    user3 = User('Siva')
    
    user1.add_to_cart(mobile,2)
    user2.add_to_cart(laptop,3)
    user1.add_to_cart(laptop,1)
    user3.add_to_cart(laptop,1)
    
    user1.view_cart()
    user2.view_cart()
    user3.view_cart()
    
    user1.save_cart()
    user2.save_cart()
```
